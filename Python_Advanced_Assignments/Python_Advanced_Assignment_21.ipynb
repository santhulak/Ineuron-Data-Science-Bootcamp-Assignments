Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?





In statistics, a probability distribution is a function that maps each possible outcome of a random variable to its probability of occurrence. In other words, it specifies the probability of observing different outcomes in a random experiment.

While the values that a random variable can take on are not known beforehand, the probabilities of those values can be calculated based on the distribution. This allows us to make predictions about the likelihood of different outcomes, even though we cannot know the precise outcome in advance. 

Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?





Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated from physical sources of randomness, such as atmospheric noise, radioactive decay, or thermal noise, and they are truly unpredictable. Pseudo-random numbers, on the other hand, are generated by deterministic algorithms that use a starting value called a seed. While the output of a good pseudo-random number generator appears to be random, it is actually deterministic, meaning that the same seed will always produce the same sequence of numbers.

The reason why pseudo-random numbers are considered "good enough" is that they can be used in many applications where true randomness is not necessary, such as simulations, games, and cryptography. Pseudo-random numbers can be generated quickly and efficiently, and their statistical properties can be analyzed to ensure that they behave like truly random numbers.

Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?





The two main factors that influence the behavior of a normal probability distribution are the mean (average) and the standard deviation of the data set. The mean determines the center or peak of the distribution, while the standard deviation indicates the spread or dispersion of the data around the mean.

Q4. Provide a real-life example of a normal distribution.





A real-life example of a normal distribution is the distribution of human height. In most populations, heights tend to follow a normal distribution, with the majority of people falling somewhere near the middle of the distribution, and fewer people at the extremes of very tall or very short heights. This distribution can be described by a mean and standard deviation, which can be used to make predictions about the proportion of people falling into different height categories.

In the short term, it is possible that a probability distribution may not behave as expected due to the element of randomness involved. However, as the number of trials grows, the behavior of the probability distribution will tend to converge to its expected behavior as defined by the underlying probability model. This is known as the law of large numbers, which states that as the number of trials increases. Therefore, in the long run, the behavior of a probability distribution can be predicted with greater accuracy.

Q8. What is the relationship between exponentiation and logarithms?





Exponentiation and logarithms are inverse operations of each other. If we take a number x and raise it to a power of y, we get a result z

z = x^y

If we know the values of x and z, and we want to find the value of y, we can use logarithms. Specifically, if we take the logarithm of z with base x, we get the value of y:

y = log(x)(z)

Q9. What are the three logarithmic functions that Python supports?

Python's math module supports three logarithmic functions:

math.log(x[, base]): This function returns the natural logarithm (base e) of x, or the logarithm of x to the specified base. If base is not provided, it defaults to e.

math.log10(x): This function returns the base-10 logarithm of x.

math.log2(x): This function returns the base-2 logarithm of x.
